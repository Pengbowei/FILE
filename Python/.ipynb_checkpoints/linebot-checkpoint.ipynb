```python
import requests
import sqlite3 as lite
import pandas as pd
from pandas.core.frame import DataFrame

def getData(year):
    payload = {
    '__EVENTTARGET': 'drpyear',
    '__EVENTARGUMENT':'',
    '__LASTFOCUS': '',
    '__VIEWSTATE': '/wEPDwULLTE3NzUyMTgxMTBkZH2SH/sUVlI18uJEZqbnmaqVc7Yl',
    '__VIEWSTATEGENERATOR':' C7C932D3',
    '__EVENTVALIDATION': '/wEdAAmq8yiJs84h1Dm1DpbAG3Mw+4UF/bLykuCuSSh3sfGOmEAHWVh8ShJorSzlMgtl+NYkHjlNrHgKY3qTCGRfPq47oaEkNE6rstEi3W2UvsMSweNlMzhYbeYKzOh8pkgKIWuNU0Dvc6Cv5ghvRvLSclInRbI1baqxHvm9HjK8Mt3LBb/tM9z6okju5ncrD6Fv5GuiQzHb5Yx1HgfOt0vEsiOvwEE7aA==',
    'drpyear': year
    }
    
    url = requests.post('https://www.findcar.com.tw/oil_w.aspx' , data = payload)
    DATA = pd.read_html(url.text)
    return DATA

def getDate(DATA):
    t = ''
    count = 0
    DATA = DATA[0]
    DATA = DATA['施行日期']
    DATA = DATA['施行日期'].tolist()

    for i in range(len(DATA) -1 , 0 , -1):
        count += 1
        t += '第' + str(count) + '周 :'
        t += str(DATA[i]) 
        if(i != 1):
            t += '\n\n'
    return t

def getPrice(DATA , SEARCH):
    try:
        SEARCH = int(SEARCH)
        DATA = DATA[0]
        PRICE92 = DATA[('92無鉛汽油', '中油')].tolist()
        PRICE95 = DATA[('95無鉛汽油', '中油')].tolist()
        PRICE98 = DATA[('98無鉛汽油', '中油')].tolist()
        PRICE10 = DATA[('高級柴油', '中油')].tolist()
        PRICE92.reverse()
        PRICE95.reverse()
        PRICE98.reverse()
        PRICE10.reverse()
        
        REPLY = '為您查詢到的油價如下\n92無鉛汽油: ' + str(PRICE92[SEARCH - 1]) + ' NTD\n'
        REPLY += '95無鉛汽油: ' + str(PRICE95[SEARCH - 1]) + ' NTD\n'
        REPLY += '98無鉛汽油: ' + str(PRICE98[SEARCH - 1]) + ' NTD\n'
        REPLY += '高級柴油: ' + str(PRICE10[SEARCH - 1]) + ' NTD'
        
        return REPLY
        
    except:
        WARNING = '無法查詢到 "' + str(SEARCH) +'" 的相關訊息\n麻煩重新輸入正確的週數'
        return WARNING

def getNowPrice():
    url = requests.post('https://www.findcar.com.tw/oil_w.aspx')
    DATA = pd.read_html(url.text)
    
    DATA = DATA[0]
    PRICE92 = DATA[('92無鉛汽油', '中油')].tolist()
    PRICE95 = DATA[('95無鉛汽油', '中油')].tolist()
    PRICE98 = DATA[('98無鉛汽油', '中油')].tolist()
    PRICE10 = DATA[('高級柴油', '中油')].tolist()
    
    REPLY = '為您查詢到的油價如下\n92無鉛汽油: ' + str(PRICE92[0]) + ' NTD\n'
    REPLY += '95無鉛汽油: ' + str(PRICE95[0]) + ' NTD\n'
    REPLY += '98無鉛汽油: ' + str(PRICE98[0]) + ' NTD\n'
    REPLY += '高級柴油: ' + str(PRICE10[0]) + ' NTD'
    
    return REPLY

def getNextWeek():
    url = requests.get('https://toolbxs.com/zh-TW/detector/gasoline_price/next_week')
    DATA = pd.read_html(url.text)
    DATA = DATA[0]
    PRICE = DATA['預測下週價格'].tolist()
    MOVE = DATA['預測變動'].tolist()

    REPLY = '為您查詢到的油價如下\n92無鉛汽油: ' + str(PRICE[0])
    REPLY += ' ; 下週調整價格為 : ' + str(MOVE[0]) + '\n\n'
    REPLY += '95無鉛汽油: ' + str(PRICE[1]) 
    REPLY += ' ; 下週調整價格為 : ' + str(MOVE[1]) + '\n\n'
    REPLY += '98無鉛汽油: ' + str(PRICE[2])
    REPLY += ' ; 下週調整價格為 : ' + str(MOVE[2]) + '\n\n'
    REPLY += '高級柴油: ' + str(PRICE[3])
    REPLY += ' ; 下週調整價格為 : ' + str(MOVE[3]) + ''
    
    return REPLY
```
